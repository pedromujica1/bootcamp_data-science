# Numpy

## Criação de Arrays

```python
# Criando um array unidimensional
arr = np.array([1, 2, 3, 4])

# Criando um array bidimensional (matriz)
matriz = np.array([[1, 2], [3, 4]])

# Criando um array com valores aleatórios
rand_arr = np.random.rand(3, 3)  # 3x3 array com valores aleatórios entre 0 e 1

# Criando um array de zeros
zeros_arr = np.zeros((2, 3))

# Criando um array de uns
ones_arr = np.ones((2, 2))

# Exibindo os arrays
print("Array:", arr)
print("Matriz:\n", matriz)
print("Array aleatório:\n", rand_arr)
print("Array de zeros:\n", zeros_arr)
print("Array de uns:\n", ones_arr)
```

## Atributos de Arrays
Cada array NumPy possui vários atributos que fornecem informações úteis sobre a estrutura dos dados:

- shape: Mostra as dimensões do array.
- dtype: Tipo de dado armazenado (int, float, etc.).
- ndim: Número de dimensões do array.
- size: Número total de elementos no array.

```python
matriz = np.array([[1, 2, 3], [4, 5, 6]])
print("Dimensões do array:", matriz.shape)
print("Tipo de dado:", matriz.dtype)
print("Número de dimensões:", matriz.ndim)
print("Número total de elementos:", matriz.size)
```

## Operações de Arrays

```python
# Soma de arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = a + b

# Subtração de arrays
d = a - b

# Multiplicação de arrays (elemento a elemento)
e = a * b

# Divisão de arrays (elemento a elemento)
f = a / b

#Multiplicação escolar (elemento a elemento)
g = a * 2

# Exibindo os resultados
print("Soma:", c)
print("Subtração:", d)
print("Multiplicação:", e)
print("Multiplicação por 2:", g)
print("Divisão:", f)
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
c = np.dot(a, b)  # Multiplicação de matrizes
print(c)
```

### Algumas funções matematicas com Arrays

```python
matriz = np.array([[1, 2, 3], [4, 5, 6]])
# Soma total dos elementos
soma_total = np.sum(matriz)

# Média dos elementos
media = np.mean(matriz)

# Soma ao longo de um eixo
soma_por_colunas = np.sum(matriz, axis=0)

print("Soma total:", soma_total)
print("Média:", media)
print("Soma por colunas:", soma_por_colunas)
```

#### Funções estatisticas

```python
matriz = np.array([[1, 2, 3], [4, 5, 6]])
# Desvio padrão
desvio_padrao = np.std(matriz)

# Mediana
mediana = np.median(matriz)

# Valor máximo
maximo = np.max(matriz)

# Valor mínimo
minimo = np.min(matriz)

print("Desvio padrão:", desvio_padrao)
print("Mediana:", mediana)
print("Valor máximo:", maximo)
print("Valor mínimo:", minimo)
```

```python
# Exponencial de um array
a = np.array([1, 2, 3])
exponencial = np.exp(a)
print(exponencial)
```

## Manipulação de Arrays

#### Geração de valores

```python
# Gerar valores aleatorios
aleatorios = np.random.rand(3, 3)  # 3x3 array com valores aleatórios entre 0 e 1
print(aleatorios)

# Gerar valores fixos
fixos = np.full((2, 2), 7)  # 2x2 array com todos os valores iguais a 7
print(fixos)

# Gerar valores em uma faixa
faixa = np.arange(0, 10, 2)  # Array com valores de 0 a 10 (exclusivo) com passo de 2
print(faixa)

# Gerar valores igualmente espaçados
espaçados = np.linspace(0, 1, 5)  # 5 valores igualmente espaçados entre 0 e 1
print(espaçados)
```

```python
matriz = np.array([[1, 2, 3], [4, 5, 6]])
# Acessando elementos
elemento = matriz[1, 1]  # Acessa o elemento da linha 2, coluna 2
print("Elemento selecionado:", elemento)

# Modificando elementos
matriz[0, 1] = 10
print("Matriz modificada:\n", matriz)

# Slicing
submatriz = matriz[:2, 1:]
print("Submatriz:\n", submatriz)

# Redimensionando (reshape)
reshaped = matriz.reshape(2, 3)
print("Matriz redimensionada:\n", reshaped)

# Transposição de matrizes
transposta = matriz.T
print("Matriz transposta:\n", transposta)

# Concatenando arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
concatenado = np.concatenate((a, b))
print("Array concatenado:", concatenado)
```

### Manipulação de Dados Ausentes

```python
# Criando um array com um valor ausente (NaN)
dados = np.array([1, 2, np.nan, 4, 5])

# Verificando onde os dados estão ausentes
mascara_nan = np.isnan(dados)

# Substituindo valores ausentes por um valor específico
dados_limpos = np.where(np.isnan(dados), 0, dados)
print("Dados com NaN substituído por 0:", dados_limpos)
```

### Indexação Booleana e Máscaras

```python
# Array de dados
dados = np.array([1, 2, 3, 4, 5])

# Selecionando elementos maiores que 3
selecionados = dados[dados > 3]
print("Elementos maiores que 3:", selecionados)
```

### Ordenação de Array

```python
# Ordenando um array
arr_desordenado = np.array([3, 1, 2, 5, 4])
arr_ordenado = np.sort(arr_desordenado)
print("Array ordenado:", arr_ordenado)

# Encontrando a posição de um elemento
posicao = np.searchsorted(arr_ordenado, 3)
print("Posição do número 3:", posicao)
```

###  Arrays Estruturados e Tipos de Dados Avançados

```python
# Definindo um array estruturado
dados_estruturados = np.array([(1, 'Alice', 23.5), (2, 'Bob', 25.0)],
                              dtype=[('id', 'i4'), ('nome', 'U10'), ('peso', 'f4')])

# Acessando elementos do array estruturado
print("Primeiro elemento:", dados_estruturados[0])
# Acessando uma coluna específica
print("Coluna nome:", dados_estruturados['nome'])
# Acessando um elemento específico de uma coluna
print("Nome do segundo elemento:", dados_estruturados[1]['nome'])
```

# Matplotlib

## Estrutura
A estrutura básica para criar um gráfico envolve o uso de Figure e Axes.

Figure e Axes
* Figure: Representa a janela ou a página onde o gráfico será desenhado.
* Axes: Um único gráfico ou plot em uma figura. Pode haver múltiplos Axes em uma Figure.

```python
# Criando uma figura com um único Axes
fig, ax = plt.subplots()
# Plotando dados no Axes
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Exibindo o gráfico
plt.show()
```

## Tipos comuns de graficos

### Grafico de linha
Ideal para mostrar tendências ao longo do tempo ou para visualizar a relação entre duas variáveis.
Trabalha com duas variaveis x e y que necessariamente precisam ter o mesmo tamanho

```python
# Gráfico de linha simples
x = [0, 1, 2, 3]
y = [0, 1, 4, 9]

# Definindo rótulos dos eixos
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Gráfico de Linha")  # Título do gráfico
plt.legend()  # Adicionando legenda
plt.plot(x, y, label="Linha de dados")  # Plotando os dados
plt.show()
```

### Grafico de linha mais elaborado

```python
# Dados para o gráfico
x = np.linspace(0, 10, 100)  # 100 pontos no intervalo de 0 a 10
y1 = np.sin(x)  # Definindo a função Seno para cada um dos pontos
y2 = np.cos(x)  # Definindo a função Cosseno para cada um dos pontos

# Criando a figura e os eixos
fig, ax = plt.subplots()

# Plotando as linhas
ax.plot(x, y1, label='Seno', color='blue', linestyle='--')
ax.plot(x, y2, label='Cosseno', color='red', marker='o')

# Adicionando título e rótulos dos eixos
ax.set_title('Gráfico de Seno e Cosseno')
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Adicionando grade
ax.grid(True)

# Adicionando legenda
ax.legend()

# Exibindo o gráfico
plt.show()
```

## Grafico de Barras
Usado para comparar valores entre diferentes grupos.


```python
# Dados
categorias = ['A', 'B', 'C', 'D']
valores = [3, 7, 5, 4]

plt.bar(categorias, valores)  # Variaveis x e y precisam ter o mesmo tamanho
plt.xlabel("Categorias")
plt.ylabel("Valores")
plt.title("Gráfico de Barras")
plt.show()
```

### Grafico de barras mais elaborado

```python

# Dados para o gráfico
categorias = ['Categoria A', 'Categoria B', 'Categoria C', 'Categoria D']
grupo1 = [25, 30, 22, 35]
grupo2 = [40, 25, 38, 18]
grupo3 = [15, 28, 32, 25]

# Largura das barras
largura_barra = 0.2

# Posições das barras no eixo x
posicoes_grupo1 = np.arange(len(categorias))
posicoes_grupo2 = [x + largura_barra for x in posicoes_grupo1]
posicoes_grupo3 = [x + largura_barra for x in posicoes_grupo2]

# Criando a figura e os eixos
fig, ax = plt.subplots()

# Plotando as barras para cada grupo
barras_grupo1 = ax.bar(posicoes_grupo1, grupo1, largura_barra, label='Grupo 1')
barras_grupo2 = ax.bar(posicoes_grupo2, grupo2, largura_barra, label='Grupo 2')
barras_grupo3 = ax.bar(posicoes_grupo3, grupo3, largura_barra, label='Grupo 3')

# Adicionando título e rótulos dos eixos
ax.set_title('Gráfico de Barras com Múltiplos Grupos')
ax.set_xlabel('Categorias')
ax.set_ylabel('Valores')



# Adicionando legenda
ax.legend()

# Exibindo o gráfico
plt.show()
```

## Grafico de Dispersão
Ideal para mostrar a relação entre duas variáveis contínuas.

```python
# Dados
x = np.arange(10)
y = x ** 2

# Criando o gráfico de dispersão

plt.scatter(x, y)  # Variaveis x e y precisam ter o mesmo tamanho
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Gráfico de Dispersão")
plt.show()
```

### Grafico mais elaborado

```python
# Dados para o gráfico
x = np.random.randn(100)
y = np.random.randn(100)
cores = np.random.rand(100)
tamanhos = 100 * np.random.rand(100)

# Criando a figura e os eixos
fig, ax = plt.subplots()

# Plotando os pontos
scatter = ax.scatter(x, y, c=cores, s=tamanhos, alpha=0.5, cmap='viridis')

# Adicionando título e rótulos dos eixos
ax.set_title('Gráfico de Dispersão')
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Adicionando barra de cores
plt.colorbar(scatter) # Passando o grafico para mudar adicionar a barra de calor

# Exibindo o gráfico
plt.show()
```

## Histograma
Usado para mostrar a distribuição de um conjunto de dados.

```python
# Dados
dados = np.random.randn(1000)  # Distribuição normal

plt.hist(dados, bins=4)  # Número de barras (bins)
plt.xlabel("Valores")
plt.ylabel("Frequência")
plt.title("Histograma")
plt.show()
```

### Histograma mais elaborado

```python
# Dados de exemplo (vamos gerar dados aleatórios para um histograma mais interessante)
dados = np.random.normal(0, 1, 1000)

# Criando o histograma com densidade de probabilidade
plt.hist(dados, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')

# Adicionando labels e título
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.title('Histograma Elaborado')

# Adicionando linhas de grade
plt.grid(True, linestyle='--', alpha=0.5)

# Exibindo o gráfico
plt.show()
```

# Integrando ambos com pandas

```python
import pandas as pd
```

```python
# Carregar os dados
df = pd.read_csv('dados_clima.csv', sep=';', decimal=',')
```

```python
df.head()
```

```python
# Criando um array NumPy a partir de uma coluna do DataFrame
array = np.array(df['Radiacao (KJ/m²)'])

#Substituindo NaN por 0
array = np.nan_to_num(array)

df['Radiacao (KJ/m²)'] = array

df.head()
```

```python
# Normalização de dados
# Criando um gráfico de linhas a partir da temp. max e Data
df.plot(x='Data', y='Temp. Max. (C)', kind='line',figsize=(10, 6))

# Customizando o gráfico
plt.title('Gráfico de Linhas')
plt.xlabel('Data')
plt.ylabel('Temp. Max. (C)')

plt.show()
```

```python
# Somar as temperaturas de um dia e tirar a media
mediaMax = df.groupby('Data')['Temp. Max. (C)'].mean()
mediaMin = df.groupby('Data')['Temp. Min. (C)'].mean()
# Pegar só os valores das temperaturas
mediaMax = mediaMax.values
mediaMin = mediaMin.values
media = (np.array(mediaMax) + np.array(mediaMin)) / 2

# Gerar grafico de barras
datas = df['Data'].unique()
plt.bar(datas, media)

#Adicionando as maximas e minimas
plt.plot(datas, mediaMax, color='red', linestyle='--')
plt.plot(datas, mediaMin, color='blue', linestyle='--')

# Customizando o gráfico
plt.title('Gráfico de Barras')
plt.xlabel('Data')
plt.ylabel('Temp. Max. (C)')
plt.show()
```
