# Estruturas de dados principais do Pandas
* **Series**: Um array unidimensional que pode armazenar dados de diferentes tipos (inteiros, strings, floats, etc.). Similar a uma coluna em uma tabela.
* **DataFrame**: Uma estrutura de dados bidimensional, como uma tabela com linhas e colunas.

## Series

```python
# Series
data = [10, 20, 30, 40, 50]
series = pd.Series(data) # Podem ser criadas a partir de listas
print(series)
```

```python
data = {'Quantidade': 10, 'Nome': "Banana", 'Valor': 10.99}
series = pd.Series(data) # Podem ser criadas a partir de dicionários
print(series)
```

## DataFrames

```python
# Criando um DataFrame a partir de um dicionário de listas
data = {
    'Nome': ['Alice', 'Bob', 'Charlie'],
    'Idade': [24, 27, 22],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte']
}

df = pd.DataFrame(data)
print(df)
```

## Manipulação dos dados

```python
# Adicionar uma nova coluna
df['Nova Coluna'] = 0 # Adiciona uma nova coluna com valor 0
df.head()
```

```python
# Remover uma coluna
df = df.drop('Nova Coluna', axis=1) # Remove a coluna 'Nova Coluna'
df.head()
```

```python
# Renomear colunas
df = df.rename(columns={'Temp. Max. (C)': 'Temperatura Maxima', 'Temp. Min. (C)': 'Temperatura Minima'}) # Renomeia as colunas
df.head()
```

```python
# Ordenar os dados
df_max_temperatures = df.sort_values(by='Temperatura Maxima', ascending=False) # Ordena o DataFrame pela coluna 'Temperatura Maxima' em ordem decrescente
df_max_temperatures.head()
```

```python
df_ascending = df.sort_values(by='Temperatura Maxima', ascending=True) # Ordena o DataFrame pela coluna 'Temperatura Maxima' em ordem crescente
df_ascending.head()
```

## Alguns tratamentos para dados faltantes

```python
# Verificar dados faltantes
print(df.isnull().sum())
# Remover linhas com dados faltantes
df_drop = df.dropna()
# Preencher dados faltantes
df_drop # Retirou as linhas da radiação NaN
```

```python
df = df.fillna(0) #Preenche os NaN com 0
df
```

## Outras funções importantes

### Funções estatisticas

```python
# Media de uma coluna
media = df['Temperatura Maxima'].mean()
print(f"Média da Temp Max: {media}")

# Somar os valores de uma coluna
soma = df['Temperatura Maxima'].sum()
print(f"Soma da Temperatura Maxima: {soma}")
# Mediana de uma coluna
mediana = df['Temperatura Maxima'].median()
print(f"Mediana da Temperatura Maxima: {mediana}")

# Desvio padrão de uma coluna
desvio_padrao = df['Temperatura Maxima'].std()
print(f"O Desvio padrão é: {desvio_padrao}")

# Correlação entre colunas
correlacao = df['Temperatura Maxima'].corr(df['Temperatura Minima'])
print(f"A corelação é:  {correlacao}")
```

### Mesclagem de dados

```python
#Junção de colunas
df_juncao = pd.concat([df['Temperatura Maxima'], df['Temperatura Minima']], axis=1)
df_juncao
```

### Agrupamento de dados

```python
# Agrupando dados a partir da coluna Data
df_agrupado = df.groupby('Data')['Temperatura Maxima'].max()
df_agrupado
```

### Aplicando uma função a uma nova coluna
Vamos imaginas que uma coluna "Temperatura Media" = (Temperatura Maxima + Temperatura Minima)/2

```python
df['Temperatura Media'] = df.apply(lambda row: (row['Temperatura Maxima'] + row['Temperatura Minima']) / 2, axis=1)
df.head()
```

## Graficos de barra

```python
# Primeiro vamos pegar o valor máximo de cada dia e concatenar os daddos
max_temp = df.groupby('Data')['Temperatura Maxima'].max().reset_index()
max_temp
```

```python
max_temp.plot.bar(x='Data', y='Temperatura Maxima') #No eixo X colocamos a data correpondente e no Y os valores
```

## Grafico de dispersão

```python
# Scatter a partir da data
df.plot.scatter(x='Temperatura Maxima', y='Temperatura Minima') #Visivel que a variação da tempetura máxima e minima não é muito
```

## Grafico de pizza

```python
# Acumular a radiação diaria
radiacao = df.groupby('Data')['Raj. Vento (m/s)'].sum().reset_index()
radiacao
```

```python
# Criar um gráfico de pizza, legenda será as datas
radiacao.plot.pie(y='Raj. Vento (m/s)', labels=radiacao['Data'], autopct='%1.1f%%', figsize=(8, 8))
```
